```python
def printitems(dictObj, indent=0):
    p=[]
    p.append('<ul>\n')
    for k in dictObj:
        v=dictObj[k]
        if isinstance(v, dict):
            p.append('<li>'+ k+ ':')
            p.append(printitems(v))
            p.append('</li>')
        else:
            p.append('<li>'+ k+ ':'+ str(v)+ '</li>')
    p.append('</ul>\n')
    return '\n'.join(p)
```
